```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import Image, display

mpl.rcParams['figure.figsize'] = (45,19)
mpl.rcParams['axes.grid'] = False


plt.style.use('seaborn')
```

```python
def generate_graph(a,b,mod1,mod2,mod3,mod4,mod5,mod6,test):
    model1 = mod1[:,[a,]]
    model1 = np.asarray(model1).reshape(-1)
    
    model2 = mod2[:,[a,]]
    model2 = np.asarray(model2).reshape(-1)

    model3 = mod3[:,[a,]]
    model3 = np.asarray(model3).reshape(-1)
    
    model4 = mod4[:,[a,]]
    model4 = np.asarray(model4).reshape(-1)

    model5 = mod5[:,[a,]]
    model5 = np.asarray(model5).reshape(-1)
    
    model6 = mod6[:,[a,]]
    model6 = np.asarray(model6).reshape(-1)    
    
    tests = test[:,[a,]]
    tests = np.asarray(tests).reshape(-1)
    
    fig1, ax1 = plt.subplots()
    fig2, ax2 = plt.subplots()
    fig3, ax3 = plt.subplots()
    fig4, ax4 = plt.subplots()
    fig5, ax5 = plt.subplots()
    fig6, ax6 = plt.subplots()
        
    ax1.plot(model1, color = 'green', linestyle='-', label='Decision Tree Regression')
    ax2.plot(model2, color = 'green', linestyle='-', label='K Neighbor Regression')
    ax3.plot(model3, color = 'green', linestyle='-', label='Lasso Regression')
    ax4.plot(model4, color = 'green', linestyle='-', label='LGBM Regression')
    ax5.plot(model5, color = 'green', linestyle='-', label='Linear Regression')
    ax6.plot(model6, color = 'green', linestyle='-', label='SGD Regression')    
    
    ax1.plot(tests, color = 'coral', linestyle='--', label='True Value')
    ax2.plot(tests, color = 'coral', linestyle='--', label='True Value')
    ax3.plot(tests, color = 'coral', linestyle='--', label='True Value')
    ax4.plot(tests, color = 'coral', linestyle='--', label='True Value')
    ax5.plot(tests, color = 'coral', linestyle='--', label='True Value')
    ax6.plot(tests, color = 'coral', linestyle='--', label='True Value')
    
    ax1.axis([0, 200, -10, 600])
    ax2.axis([0, 200, -10, 600])
    ax3.axis([0, 200, -10, 600])
    ax4.axis([0, 200, -10, 600])
    ax5.axis([0, 200, -10, 600])
    ax6.axis([0, 200, -10, 600])
    
    ax1.legend()
    ax1.set_title('Decision Tree Regression for '+b)
    ax2.legend()
    ax2.set_title('K Neighbor Regression for '+b)
    ax3.legend()
    ax3.set_title('Lasso Regression for '+b)
    ax4.legend()
    ax4.set_title('LGBM Regression for '+b)
    ax5.legend()
    ax5.set_title('Linear Regression for '+b)
    ax6.legend()
    ax6.set_title('SGD Regression for '+b)

#     plt.plot(model1, label='Decision Tree Regression')
#     plt.plot(model2, label='K Neighbor Regression') 
#     plt.plot(model3, label='Lasso Regression') 
#     plt.plot(model4, label='LGBM Regression') 
#     plt.plot(model5, label='Linear Regression')
#     plt.plot(model6, label='SGD Regression') 
#     plt.plot(tests, label='True Value')
    
#     plt.legend(['Decision Tree Regression', 'K Neighbor Regression', 'Lasso Regression', 'LGBM Regression', 'Linear Regression', 'SGD Regression', 'True Value'], loc='upper right')
#     plt.title('Graph for '+b+' Prediction')
    
    MAE = np.mean(np.abs(model1-tests))
    print("MAE of "+b+" for Decision Tree Regression model is: ",MAE)
    MAE = np.mean(np.abs(model2-tests))
    print("MAE of "+b+" for K Neighbor Regression model is: ",MAE)
    MAE = np.mean(np.abs(model3-tests))
    print("MAE of "+b+" for Lasso Regression model is: ",MAE)
    MAE = np.mean(np.abs(model4-tests))
    print("MAE of "+b+" for LGBM Regression model is: ",MAE)
    MAE = np.mean(np.abs(model5-tests))
    print("MAE of "+b+" for Linear Regression model is: ",MAE)
    MAE = np.mean(np.abs(model6-tests))
    print("MAE of "+b+" for SGD Regression model is: ",MAE)
```
